```python
import pandas as pd

from paths import DATA_PATH
```

```python
news_path = DATA_PATH / 'news_dataframe_labeled.csv'
stocks_path = DATA_PATH / 'stocks_dataframe.csv'
```

```python
news_df = pd.read_csv(news_path, parse_dates=True)
news_df
```

```python
stocks_df = pd.read_csv(stocks_path, parse_dates=True)
stocks_df
```

```python
df = news_df.merge(stocks_df, on=['date'], how='left')
df
```

# Заполняем новости неторговых дней котировками предыдущих дней

```python
fdf = df
cols = ['open','high','low','close','volume']
fdf[cols] = fdf[cols].fillna(method='ffill')
fdf
```

```python
fdf.to_csv(DATA_PATH / 'news_stocks.csv', index=False)
```
